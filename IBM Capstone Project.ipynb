So we have gotten 95% accuracy with this method. Lets see what the confusion matrix says.

It says that 105 of the benign breast cancer has been correctly determined by us. While 5 patients with malignant disease has been wrongly classified as benign.

It also says that 58 of the malignant cancer patients has been correctly classified, however 3 benign patients has been classified as malignant by out model.

This seems pretty good, lets calculate our sensitivity, specificity, positive predicted value(ppv) and negative predicted value(npv) for MALIGNANT prediction in our model because this is the more important diagnosis to be made. A wrongly diagnosed MALIGNANT patient can be detrimental!

```python
#sentivity for malignant
#formula is True Positive (TP) / TP + FN(false negative), in our case positive will be malignant and negative is benign

sensitivity = 58/(58+5) * 100

#specificity for malignant
#formula is TN / TN + FP

specificity = 105/(105+3) * 100

#positive predictive value (ppv) for malignant
#this means if a patient is predicted as positive, what are the probability that the patient is actually positive (in our case has malignant breast cancer)
#formula is TP / (TP/FP)
ppv = 58 / (58+3) * 100

#negative predictive value (npv) for malignant
#this means if a patient is predicted as negative, what are the probability that the patient is actually negative (in our case has benign breast cancer)
#formula is TN / (TN/FN)
npv = 105/ (105+5) * 100

print("Our model has:")
print(f"{sensitivity:.2f}% sensitivity")
print(f"{specificity:.2f}% specificity")
print(f"{ppv:.2f}% positive predictive value")
print(f"{npv:.2f}% negative predictive value")
```
